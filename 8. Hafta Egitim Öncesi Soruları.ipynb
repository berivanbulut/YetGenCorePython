# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
def urun_kaydet(ad, fiyat):
    with open("urunler.txt", "a") as dosya:
        dosya.write(ad + "," + fiyat + "\n")

    print("Ürün başarıyla kaydedildi.")


def kelime_degistir(dosya_adi, eski_kelime, yeni_kelime):
    with open(dosya_adi, 'r') as dosya:
        metin = dosya.read()
        
    metin = metin.replace(eski_kelime, yeni_kelime)
    
    with open(dosya_adi, 'w') as dosya:
        dosya.write(metin)
    
    print(f"Metin dosyasında {eski_kelime} kelimesi {yeni_kelime} ile değiştirildi.")

ad = input("Ürün adını girin: ")
fiyat = input("Ürün fiyatını girin: ")
urun_kaydet(ad, fiyat)

dosya_adi = "urunler.txt"  
eski_kelime = input("Değiştirilecek kelimeyi girin: ")
yeni_kelime = input("Yeni kelimeyi girin: ")
kelime_degistir(dosya_adi, eski_kelime, yeni_kelime)
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
class Vehicle:
    def __init__(self, name, speed, mileage):
        self.name = name
        self.speed = speed
        self.mileage = mileage

    def display_info(self):
        print("Name:", self.name, "Speed:", self.speed, "Mileage:", self.mileage)


class Bus(Vehicle):
    pass
bus = Bus("School Volvo", 180, 12)
bus.display_info()
````
